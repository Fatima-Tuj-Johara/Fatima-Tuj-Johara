```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('fake_job_postings.csv')
df_job=df.copy()
df_job.head(2)
```

```python
df_job.columns
```

```python
df_job.isnull().sum()
```

```python
df_job.dtypes
```

```python
df_job['department']=df_job['department'].fillna(df_job['department'].mode()[0])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn import feature_selection
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
```

```python
data_cat_imp = SimpleImputer(strategy='constant', fill_value='Missing')

cat_imp_feature = ['title', 'location', 'department', 'salary_range',
                   'company_profile', 'description', 'requirements',
                   'benefits', 'employment_type', 'required_experience',
                   'required_education', 'industry', 'function']

data_num_imp=SimpleImputer(strategy='constant',fill_value=0)

num_imp_feature = ['job_id','telecommuting','has_company_logo','has_questions',
                   'fraudulent']
```

```python
#Transforming into column
data_imp_trans=ColumnTransformer([
          ("data_cat_imp",data_cat_imp,cat_imp_feature),(
            "data_num_imp",data_num_imp,num_imp_feature)
          ])
#Transforming and assining the data
transformed_data=data_imp_trans.fit_transform(df_job)
transformed_data
```

```python
all_columns = cat_imp_feature + num_imp_feature
```

```python
#Transforming data into DataFrame
df_job_transformed_data=pd.DataFrame(
               transformed_data,
               columns=all_columns)
df_job_transformed_data.head(2)
```

```python
df_job_transformed_data.isna().sum()
```

```python
X_trans=df_job_transformed_data.drop("fraudulent",axis=1)
y_trans=df_job_transformed_data['fraudulent']
y_trans=y_trans.astype(int)
#shape of features and label
X_trans.shape
y_trans.shape
X_trans.columns
```

```python
X_trans
```

```python
#Instantation of One Hot Encoder for categorical data trnasformation into Numeric

one_hot=OneHotEncoder()
clf_trans=ColumnTransformer(
    [('one hot',one_hot,cat_imp_feature)],remainder='passthrough')
X_trans_fin=clf_trans.fit_transform(X_trans)
```

```python
#Splitting the data into train and test with 23% reseverd for testing and 70% for training
X_train,X_test,y_train,y_test=train_test_split(
      X_trans_fin,y_trans,test_size=.23,random_state=42)
X_train.shape,X_test.shape,y_train.shape,y_test.shape

```

```python
model_rf=RandomForestClassifier()
#Fitting the data into model
model_rf.fit(X_train,y_train)
print(f"Fake Job Random Forest Model Accuracy:{model_rf.score(X_test,y_test)*100:.2f}%")
```

```python
y_pred_rf=model_rf.predict(X_test)
y_pred_rf
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_rf))
```
